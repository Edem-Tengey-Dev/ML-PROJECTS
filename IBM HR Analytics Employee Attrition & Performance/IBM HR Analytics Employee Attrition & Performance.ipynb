```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('IBM HR Analytics Employee Attrition & Performance.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
df["Attrition"] = LabelEncoder().fit_transform(df["Attrition"])
df.head()
```

```python
df.dtypes
```

```python
df["Over18"] = LabelEncoder().fit_transform(df["Over18"])
df.head()
```

For the Over18 column it there is only 1 value so after encoding it returns only 0

```python
df["Over18"].value_counts()
```

```python
df["BusinessTravel"] = LabelEncoder().fit_transform(df["BusinessTravel"])
df.head()
```

```python
df["BusinessTravel"].value_counts()
```

```python
df.dtypes
```

```python
df[["Department", "EducationField","Gender", "JobRole", "MaritalStatus", "OverTime"]] = df[["Department", "EducationField","Gender", "JobRole", "MaritalStatus", "OverTime"]].apply(LabelEncoder().fit_transform)
df.head()
```

```python
X = df.drop(["Attrition"], axis = 1)
y = df["Attrition"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python

linear_regression = cross_val_score(LinearRegression(), X, y, cv = 10)
print(linear_regression.mean())
logreg = cross_val_score(LogisticRegression(max_iter = 1000), X, y, cv=10)
print(logreg.mean())
svm = cross_val_score(SVC(), X, y, cv=10)
print(svm.mean())
rfc = cross_val_score(RandomForestClassifier(), X, y, cv=10)
print(rfc.mean())
dtc = cross_val_score(DecisionTreeClassifier(), X, y, cv=10)
print(dtc.mean())
gnb = cross_val_score(GaussianNB(), X, y, cv=10)
print(gnb.mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```
